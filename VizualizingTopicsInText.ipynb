# Vizualizing the Topics in Text and Testing the Test Set

```python
from constants import folder_meicogsci, all_topics_names, test_files
import collections
import gensim
import string
import nltk
import os
```

```python
# prepare helping function
lem = nltk.stem.WordNetLemmatizer()
```

```python
# choose which colors to use (they get used from left to right)
all_color_codes = [91, 92, 93, 94, 95, 96, 31, 32, 33, 34, 35, 36]
```

```python
# import a model
model = gensim.models.LdaModel.load(os.path.join("models", "LDA_21"))
model.minimum_probability = 0.01
```

```python
# for each abstract, color the text based on the topics and show the most popualr topics
for filename in test_files:
    with open(os.path.join(folder_meicogsci, filename)) as f:
        data = " ".join(f.readlines()).replace("... title: ", "")
    print(filename)
    data_2 = data
    data = data.lower().replace("\n", " ")
    data_tok = nltk.tokenize.word_tokenize(data)
    data_tok = [lem.lemmatize(w) for w in data_tok]
    doc_topics, word_topics, _ = model.get_document_topics(model.id2word.doc2bow(data_tok), per_word_topics=True)
    all_current_topics = [t for t, v in doc_topics]
    word_topics_dict = dict()
    for word_id, topics in word_topics:
        word_topics_dict[word_id] = topics[0]
    get_word_dict = dict()
    for word in data_tok:
        word_id = model.id2word.doc2bow([word])
        if word_id:
            get_word_dict[word.lower()] = word_topics_dict[word_id[0][0]]
    print(doc_topics, all_current_topics)
    for i, current_topic in enumerate(doc_topics):
        if current_topic[1] > 0.1:
            print("\x1b[" + str(all_color_codes[i]) + "m" + all_topics_names[current_topic[0]] + ": " + str(current_topic[1]) + "\x1b[0m")
    for word in data_2.split(" "):
        word_test = word.lower().translate(str.maketrans('', '', string.punctuation))
        if word_test in get_word_dict:
            try:
                print("\x1b[" + str(all_color_codes[all_current_topics.index(get_word_dict[word_test])]) + "m" + word + "\x1b[0m", end=" ")
            except:
                print(word, end=" ")
        else:    
            print(word, end=" ")
    input("next?")
```
